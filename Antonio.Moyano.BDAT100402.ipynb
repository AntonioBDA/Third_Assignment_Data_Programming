```python
#Step 1. Import the necessary libraries
import pandas as pd
#Step 2. Import the dataset from this address.
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
#Step 3. Assign it to a variable called users
users = pd.read_csv(url, index_col="user_id", sep = '|')
users.head()
```

```python
#Step 4. Discover what is the mean age per occupation
users.groupby(['occupation']).mean()
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
import numpy as np
occGenre = pd.pivot_table(users,'age','occupation','gender', aggfunc='count')
occGenre['Male Ratio'] = occGenre['F']/occGenre['M']
occGenre['Male Ratio']
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages
Ages = pd.pivot_table(users,'age','occupation',aggfunc='count')
Ages['minAge'] =users.groupby(['occupation'])['age'].min()
Ages['maxAge'] =users.groupby(['occupation'])['age'].max()
Ages.drop(('age'), inplace=True, axis=1)
Ages
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age
Ages = pd.pivot_table(users,'age','occupation','gender',aggfunc='mean')
Ages
```

```python
#Step 8. For each occupation present the percentage of women and men
percentage = pd.pivot_table(users,'age','occupation','gender', aggfunc='count')
percentage['% Women'] = percentage['F']/(percentage['F'] + percentage['M'])
percentage['% Men'] = percentage['M']/(percentage['F'] + percentage['M'])
percentage
```

<h2>Question 6</h2>
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
%matplotlib inline
USmd = pd.read_csv('us-marriages-divorces-1867-2014.csv', sep=',')
USmd["Mxcap"] = USmd["Marriages"]/USmd["Population"]
USmd["Dxcap"] = USmd["Divorces"]/USmd["Population"]
x_data = USmd["Year"]
fig, ax = plt.subplots()
ax.plot(x_data, USmd["Dxcap"])
ax.plot(x_data, USmd["Mxcap"])
ax.set_title('Marriage and Divorces per capita 1867 - 2014')
ax.set_xlabel('Year')
ax.set_ylabel('Marriage and Divorces per capita')
ax.legend(['Divorces per capita', 'Marriages per capita'])
```
